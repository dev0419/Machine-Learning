```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Read the CSV file
df = pd.read_csv('C:\\Users\\rsurs\\OneDrive\\Documents\\Machine Learning\\Data_Preprocessing\\titanic.csv')

# Define categorical features and the target variable
categorical_features = ['Sex', 'Embarked', 'Pclass']

# Handle missing values in categorical columns
imputer = SimpleImputer(strategy='most_frequent')
df[categorical_features] = imputer.fit_transform(df[categorical_features])

# Apply ColumnTransformer with OneHotEncoder
ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), categorical_features)
    ], remainder='passthrough')

X = ct.fit_transform(df)

# Convert X to a numpy array (not necessary if you want to keep it as a DataFrame)
X = np.array(X)

# Encode the target variable using LabelEncoder
le = LabelEncoder()
y = le.fit_transform(df['Survived'])

print("Updated matrix of features: \n", X)
print("Updated dependent variable vector: \n", y)
```
